```python
df = pd.read_csv('hoteis.csv', sep= ';')
df
```

```python
df.info()
```

```python
# Renomear as colunas com nomes em portugues
df.rename(columns=
      {'Rank':'Ranking',
        'Name': 'Nome',
        'Location': 'Localização',
        'Overview': 'Visão Geral',
        'Total Rooms': 'Total de Quartos',
        'Starting Rate in ($)': 'Taxa Inicial',
        'Dining Area': 'Area de Jantar',
        'Drinking Area': 'Area de Bebida',
        'Hotel Ammenties': 'Comodidades do hotel',
        'Address': 'Endereço',
        'Number': 'Telefone'}, inplace=True)
```

```python
df
```

```python
# Buscando o maior valor da coluna Taxa Inicial

maior_taxa = df['Taxa Inicial'].max()

# Buscar a linha correspondente
indice_maior_valor = df['Taxa Inicial'].idxmax()

# Exibir a linha correspondente
linha_maior_valor = df.loc[indice_maior_valor]
```

```python
print(linha_maior_valor)
```

```python
# Exibindo lista de hotéis do maior valor para o menor

df_ordenado = df.sort_values(by='Taxa Inicial',ascending=False)
df_ordenado
```

```python
# Criando um gráfico de barras com o nome dos hoteis e o valor da Taxa Inicial


plt.figure(figsize=(20,10)) # Definindo o tamanho do gráfico

plt.bar(df['Nome'], df['Taxa Inicial'], color='skyblue') # criando as barras e as cores


plt.title('Valores por Hotel', fontsize=16) # Definindo o título do gráfico
plt.xlabel('Hotel', fontsize=14) # Definindo o nome do eixo x
plt.ylabel('Taxa Inicial', fontsize=14) # Definindo o nome do eixo y


plt.xticks(rotation=45, ha='right')

plt.tight_layout()



plt.show() # Exibir o gráfico
```

```python
# Criando um gráfico de colunas com o total de quartos de cada hotel

plt.figure(figsize=(20,10)) #Definindo o tamanho do gráfico

plt.barh(df['Nome'], df['Total de Quatros'], color='orange') #criando as barras e as cores

plt.title('Quantidade de Quartos por Hotel', fontsize=16) # Título do gráfico
plt.xlabel('Hotel', fontsize=14)
plt.ylabel('Total de Quartos', fontsize=14)

plt.xticks(rotation=45, ha='right')

plt.tight_layout()

plt.show()


```

```python
# Exibir a quantidade de hoteis em cada cidade

# Agrupar por cidade e contar a quantidade de hoteis em cada uma
cidades_mais_hoteis = df.groupby('Localização').size().reset_index(name='Quantidade de Hoteis por Cidade')

# Ordenar a quantidade de hoteis do maior para o menor usando o nome correto da coluna
cidades_mais_hoteis = cidades_mais_hoteis.sort_values(by='Quantidade de Hoteis por Cidade', ascending=False)

print(cidades_mais_hoteis)
```

```python
# Criando gráficos de linhas de maior quantidade de hoteis por cidade

plt.figure(figsize=(20,10))

plt.bar(cidades_mais_hoteis['Localização'], cidades_mais_hoteis['Quantidade de Hoteis por Cidade'], color='Purple') # Criando as barras

# Adicionar rótulos e títulos
plt.title('Quantidade de Hoteis por Cidade', fontsize=16)
plt.xlabel('Cidade', fontsize=16)
plt.ylabel('Quantidade de Hoteis', fontsize=16)


plt.xticks(rotation=45)  # Rotaciona os rótulos do eixo X para melhorar a legibilidade
plt.grid()

# Exibir o gráfico
plt.tight_layout()
plt.show()


```
